```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
df= pd.read_csv('Churn_Modelling.csv')
df
```

PRE PROCESS THE DATA

```python
df.drop(columns=['RowNumber','CustomerId','Surname'], inplace= True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
label_encoder_gender= LabelEncoder()
df['Gender']= label_encoder_gender.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
onehot_encoder_country= OneHotEncoder()
country_encoder= onehot_encoder_country.fit_transform(df[['Geography']])
country_encoder
```

```python
onehot_encoder_country.get_feature_names_out()
```

```python
country_encoder.toarray()
```

```python
country_encoding= pd.DataFrame(country_encoder.toarray(),columns= onehot_encoder_country.get_feature_names_out())
country_encoding
```

```python
df
```

```python
df_new= pd.concat([df.drop(columns=['Geography']),country_encoding], axis=1)
df_new.head()
```

```python
#save the encoders

with open('label_encoder_gender.pkl', 'wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('onehot_encoder_country.pkl','wb') as file:
    pickle.dump(onehot_encoder_country, file)
```

```python
df['Exited'].value_counts()
```

```python
X= df_new.drop(columns='Exited')
y= df_new[['Exited']]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, train_size=0.8, random_state= 30)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train
```

```python
scaler= StandardScaler()
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

## ANN IMPLEMENTATION

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model= Sequential([
Dense(64,activation='relu',input_shape=(X_train_scaled.shape[1],)), #first hidden layer
Dense(32, activation='relu'), # second hidden layer
Dense(1, activation='sigmoid') # output layer
]   
)
```

```python
model.summary()
```

```python
#you can define optimisers here and can use in compile function or else you can directly use in compile func as next cell
import tensorflow
optimiser= tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss = tensorflow.keras.losses.BinaryCrossentropy()
```

```python
## compile the model
model.compile(optimizer= optimiser, loss="binary_crossentropy", metrics=['accuracy'])
```

```python
# set up the Tensorboard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
log_directory= "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback= TensorBoard(log_dir= log_directory, histogram_freq=1)
```

```python
# Early Stopping
early_stopping_callback= EarlyStopping(monitor='val_loss', patience=10,restore_best_weights=True)
```

```python
my_model= model.fit(X_train_scaled,
                    y_train,
                    validation_data= (X_test_scaled, y_test),
                    epochs=100,
                    callbacks= [tensorflow_callback,early_stopping_callback]
                   )
```

```python
model.save('model.h5')
```
